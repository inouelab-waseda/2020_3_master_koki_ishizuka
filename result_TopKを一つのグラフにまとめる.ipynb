# TopKの遷移を一つのグラフにまとめます
- 各モデルで出力したtxtから読み込む

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
category = 'Dress'
```

```python
f = open('./pickle/{}/test_pairs.pickle'.format(category),'rb')
test_pairs = pickle.load(f)
```

```python
f = open('./model/Dress/Single_normal/a0.1/512/topk_vals99.txt','rb')
# f = open('./model/T_Shirt/Single/a0.1/512/TopK_99.txt','rb')
single_model = pickle.load(f)
```

```python
f = open('./model/Dress/Single_improved/a0.1b0.05/512/topk_vals99.txt','rb')
# f = open('./model/T_Shirt/Single/a0.1b0.05/512/TopK_99.txt','rb')
singleImproved_model = pickle.load(f)
```

```python
f = open('./model/Dress/Dual_normal/a0.1/512/topK_vals.txt','rb')
# f = open('./model/T_Shirt/tripletloss/a0.1/512/topK_vals.txt','rb')
dualNormal_model = pickle.load(f)
```

```python
f = open('./model/Dress/Dual_improved/a0.1b0.05/300/topK_vals.txt','rb')
# f = open('./model/T_Shirt/improved_tripletloss/a0.1b0.05/300/topK_vals.txt','rb')
dualImproved_model = pickle.load(f)
```

```python
x = list(range(1,51))
chancelv=np.array(x)/len(test_pairs)
plt.plot(x,single_model,c='b',label='Single_M0.1D512({})'.format(round(single_model[19],3)))
plt.plot(x,dualNormal_model,c='r',label='Dual_M0.1D512({})'.format(round(dualNormal_model[19],3)))
plt.plot(x,dualImproved_model,c='g',label='DualImp_M0.1_M0.05_D300({})'.format(round(dualImproved_model[19],3)))
plt.plot(x,singleImproved_model[:50],c='m',label='SingleImp_M0.1_M0.05_D300({})'.format(round(singleImproved_model[19],3)))

# plt.plot(x,topk_vals,c='r',marker='o',label='improvedTripletLoss({})'.format(round(topk_vals[4],3)))
plt.plot(x,chancelv,linestyle='dashed',label='chance level(0.00984)')

# graph settings
plt.legend(loc='best')#,bbox_to_anchor=(1, 0.2))
plt.grid('True')
plt.title("top-k accuracy")
plt.xlabel("k-value")
plt.ylabel("top-k acc")

plt.savefig('./compare_topK.png',dpi=300)
```
